```python
import matplotlib.pyplot as plt
```

```python
region_counts = data['Region'].value_counts()
region_counts.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Bar Plot of Region Counts')
plt.xticks(rotation=45)
plt.show()
```

```python
income_counts = data['IncomeGroup'].value_counts()

income_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Pie Chart of Income Groups')

plt.axis('equal')
plt.show()
```

```python
value_counts = data['IncomeGroup'].value_counts()
value_counts.plot(kind='bar')

plt.xlabel('Income Group')
plt.ylabel('Frequency')
plt.title('Histogram of Income Groups')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot_table = data.pivot_table(index='Region', columns='IncomeGroup', aggfunc='size', fill_value=0)
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Stacked Bar Plot of Income Groups by Region')
plt.xticks(rotation=45)
plt.show()
```
